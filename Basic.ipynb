# Các phép tính

```python
tensor_1 = torch.rand(3,3,3)
tensor_2 = torch.rand(3,3,3)
print('Phep cong:', tensor_1+tensor_2)
# Không có phép nhân cho 2 khối tensor 3D,
#nên bmm hoạt động giống việc tách ra 3 ma trận 3x3 và nhân lần lượt gọi là batch matrix multiply
print('Phep nhan:', torch.bmm(tensor_1, tensor_2))
```

```python
tensor_3 = torch.rand(2,4)
# Cú pháp .transpose(input, dimension0, dimension1) # vị trí 2 dimension mà bạn muốn swap
tensor_4 = torch.transpose(tensor_3, 0, 1)
print('Truoc khi transpose:', tensor_3)
print('Sau khi transpose', tensor_4)
```

# Indexing

# Describing tensor

```python
tensor_6 = torch.rand(3, 2)
print('Shape:', tensor_6.shape)
print('Datatype:', tensor_6.dtype)
print('Number of dimensions:', tensor_6.ndim)
print('Number of elements:', tensor_6.numel())
```

# Reshape

```python
tensor_7 = torch.rand(2,3,4)
print(tensor_7)
print(tensor_7.reshape(6,4))
print(tensor_7.reshape(2, -1))
print('Tensor 1D flatten:\n', tensor_7.flatten())
```

# Unsqueeze và squeeze

```python
tensor_8 = torch.rand(2,3)
#Unsqueeze nhằm thêm một dimension mới có size là 1 vào index dimension chỉ định
# Ví dụ khởi tạo shape ban đầu là (2,3) unsqueeze ở vị trí 0 -> shape mới: (1,2,3)
tensor_9 = tensor_8.unsqueeze(0)
tensor_10 = tensor_8.unsqueeze(2)
print('Shape initially:', tensor_8.shape)
print('Shape afterwards:', tensor_9.shape)
print('Shape afterwards:', tensor_10.shape)
```

```python
# Squeeze sẽ remove những chiều có kích thước là 1
tensor_a = torch.rand(3,1,2,1)
print('Shape initially:', tensor_a.shape)
print('Shape afterwards:', tensor_a.squeeze(1).shape)
print('Shape afterwards:', tensor_a.squeeze().shape) #Remove tất cả dimension có size=1
```

# Broadcast
